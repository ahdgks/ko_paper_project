```python
# 기본 라이브러리 임포트
import pandas as pd
```

```python
total_df= pd.read_csv('paper_dataset.csv',encoding='utf-8',lineterminator='\n',index_col=0)
```

```python
total_df.dropna(inplace=True)
total_df.reset_index(inplace=True)
total_df=total_df.drop('index',axis=1)
```

# 전처리

```python
import pandas as pd
import re

def remove_rows_with_chinese(df, columns):
    for column in columns:
        df = df[~df[column].apply(lambda x: bool(re.search('[一-龥]', str(x))))]
    return df

columns = ['title', 'entire_orginal_text', 'entire_summary_text', 'section_orginal_text', 'section_summary_text']
total_df = remove_rows_with_chinese(total_df, columns)

#사람이름 3글자 단위로 끊기게 전처리
def shorten_authors(authors):
    author_list = authors.split(',')
    shortened_authors = [author.strip()[:3] for author in author_list]
    return ', '.join(shortened_authors)

total_df['author'] = [re.sub('[^가-힣,]', '', s) for s in total_df['author']]
total_df['author'] = total_df['author'].apply(shorten_authors)
total_df.dropna(inplace=True)
total_df.reset_index(inplace=True)
total_df=total_df.drop('index',axis=1)
```

```python
#rag에 들어갈 텍스트 만들기(section_summary_text)
for i in range(len(total_df)):
    total_df.loc[i,'rag_text']= '이 논문의 제목은 "{0}"이다. 저자는 "{2}"이다. 발행연도는 "{3}"이다.논문의 내용은 {1} 이다.'.format(total_df.loc[i,'title'],total_df.loc[i,'section_summary_text'],total_df.loc[i,'author'],total_df.loc[i,'date'])
```

```python
a=[]
for index,length in enumerate(total_df['rag_text']):
    a.append((index,len(length)))
    
sorted_a=sorted(a,key=lambda x: x[1],reverse=True)
sorted_a
```

```python
#너무 긴 텍스트 제거 
total_df.drop(sorted_a[0][0],inplace=True)
total_df.reset_index(inplace=True)
total_df=total_df.drop('index',axis=1)
```

```python
df=total_df[['doc_id','title','date','ipc','rag_text']]


df.to_csv('preprocessing_dataset.csv',encoding='utf-8')
```
