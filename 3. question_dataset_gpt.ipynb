```python
import pandas as pd
data=pd.read_csv('preprocessing_dataset.csv',index_col=0)
```

```python
data
```

```python
from openai import OpenAI

client = OpenAI(
  api_key= "your_key"
)

gpt_text=[]
for s in data[data['ipc']=='사회과학']['rag_text'][-1900:]:
    completion = client.chat.completions.create(
    model="gpt-4-0125-preview",
    messages=[
        {"role": "system", "content": """너는 논문 검색 데이터셋 생성기야. 아주 중요해.  
        \n 1. Q&A형식으로 대답해줘.  2.주어진 ###의 "핵심 단어 두개만" 사용해서 관련된 논문을 찾아달라는 질문을 만들어줘. 
        말투는 ~~ 에 관한 논문을 찾아줘. 라고 해 3.A는 관련된 논문은 이라는 말로 시작해줘. 
        4. A는 제목,저자,발행연도, 내용이 들어간 문장형식으로 만들어줘.5.내용 파트를 더 길게 만들어줘."""},
        {"role": "user", "content": "###:"+s }
            ]
                                                )
    print(completion.choices[0].message.content)
    gpt_text.append(completion.choices[0].message)
```

```python
qa_set=[]
for i in range(len(gpt_text)):
    qa_set.append(gpt_text[i].content)
    
question=[]
answer=[]
for i in range(len(qa_set)):
    qa_set[i]=qa_set[i].replace('\n\n','\n')
    a=qa_set[i].split('\n')
    question.append(a[0])
    answer.append(a[1])
    
for i in range(len(question)):
    question[i]=question[i].replace('"','')
    question[i]=question[i].replace('Q: ','')
    
    
for i in range(len(answer)):
    answer[i]=answer[i].replace('A: ','')
```

```python
df=pd.DataFrame({'질문':question,'답변':answer})
```

```python
df.to_csv('gpt_dataset.csv')
```
