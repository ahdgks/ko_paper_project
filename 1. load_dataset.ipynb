# 데이터프레임으로 변환

```python
def make_df(file_path):
    data = json.loads(Path(file_path).read_text())
    
    # 데이터를 딕셔너리 형태로 저장하여 DataFrame 생성
    df_data = {
        'doc_id': [],
        'title': [],
        'date': [],
        'author': [],
        'ipc': [],
        'entire_orginal_text': [],
        'entire_summary_text': [],
        'section_orginal_text': [],
        'section_summary_text': []
    }

    for item in data[0]['data']:
        df_data['doc_id'].append(item['doc_id'])
        df_data['title'].append(item['title'])
        df_data['date'].append(item['date'])
        df_data['author'].append(item['author'])
        df_data['ipc'].append(item['ipc'])
        df_data['entire_orginal_text'].append(item['summary_entire'][0]['orginal_text'])
        df_data['entire_summary_text'].append(item['summary_entire'][0]['summary_text'])
        df_data['section_orginal_text'].append(item['summary_section'][0]['orginal_text'])
        df_data['section_summary_text'].append(item['summary_section'][0]['summary_text'])

    # DataFrame 생성
    df = pd.DataFrame(df_data)
    return df
```

```python
data1=make_df(file_path1)
data2=make_df(file_path2)
data3=make_df(file_path3)
data4=make_df(file_path4)
data5=make_df(file_path5)
print(len(data1),len(data2),len(data3),len(data4),len(data5))
```

```python
total_df=pd.concat([data1,data2,data3,data4,data5])
```

# csv로 변환

```python
total_df = total_df.reset_index(drop=True)
total_df.to_csv('paper_dataset.csv',encoding='utf-8')
```
